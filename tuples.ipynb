# Demetler (tuples)
Bir tuple için kısaca elemanları sonradan değiştirilemeyen liste diyebiliriz.
Tuple yapılarında köşeli parantez [] yerine, normal parantez **()** kullanılır.
Örneğin;

```
ornek_tuple = (1 ,2 ,3, 4, 5)
```
Peki o zaman neden liste yerine tuple kullanalım diye sorabilirsiniz. Bir tuple kullanmanın en büyük avantajı, zaman ve bellek verimliliğidir. Bir tuple kullandığımızda, bilgisayarın hafızasında ( yani Ram'de) arka arkaya sıralı bir yer ayrılır.Screenshot (683).png

Lakin, bir liste için sıralı bir alan ayrılmaz. Listenin içindeki herbir element için hafızada rastgele bir alan ayrılır. Ayrılan herbir alanda hem istenilen veri hem de bir sonraki elemanın hafızada depolandığı yer tutulmak zorundadır. Yukarıda gördüğünüz resimde bir liste yapısını basitleştirilmiş haliyle görüyorsunuz: üstteki kırmızılar adres bilgisini yani bloğun şu anda nerede olduğunu gösteriyor. Ortadaki gri alan asıl veriyi tutuyor. Alttaki yeşil alan da bir sonraki elementin adres bilgisini tutuyor.
Fazladan tutulan adres bilgileri de hem fazladan yer kaplanmasına hem de programın yavaş çalışmasına sebep oluyor.
Bu sebeplerden dolayı, eğer içindeki elemanlar bir kere tanımlanıp bir daha değişmeyecekse tuple kullanmak daha iyi olacaktır.

````python
t = (1,2,"Huy")
t = list(t)
t[0] = 3
t = tuple(t)
t
````
