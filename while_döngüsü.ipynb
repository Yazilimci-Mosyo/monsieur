# While Döngüsü
While döngüleri, bir veya daha fazla kod satırının belirli bir koşul sağlandığı sürece arka arkaya çalıştırılmasını sağlar. 

Yazımı if koşullu ifadelerine benzer:
```
While koşul:     
    print("Merhaba")
    print("Dunya")
```

başlangıca while kelimesi konarak while döngüsünde olunduğu belirtilir.
arkasından koşul belirtilir ve koşulun arkasından iki nokta işareti konur.
Bir alt satıra geçilir ve 2 veya 4 boşluk bırakılarak alt alta kodlar yazılır.

Görüldüğü üzere yazım olarak if koşullu ifadelerinden tek farkı, ifadenin başına while yazılmasıdır.


```
if koşul:
    print("merhaba")
```


```
while koşul:
    print("merhaba")
```




##Örnek: Yanlış cevap verildiği sürece tekrar ve tekrar sorulan quiz sorusu

While döngüsü kullanımına aşağıdaki gibi bir örnek verilebilir:

bu örnekte doğru cevap verilmediği sürece soru tekrar ve tekrar sorulmaktadır

````python
dogrucevapverildi = False
 
while dogrucevapverildi is False:
    cevap = input("3 + 2 = ?\n")

    if cevap is "5":
      print("Doğru!")
      dogrucevapverildi = True
    else:
      print("yanlış cevap, lütfen tekrar deneyiniz.")

print("döngüden çıktım.")
````

## Break komutu
Bazen, başlangıçta döngünün duracağı koşulu vermek yerine döngü içinde belirli bir durum yaşandığında döngüden çıkmak isteyebiliriz. Böyle durumlarda break komutunu kullanabiliriz. Break komutu,döngüyü anında sonlandırmaya yarar.
Yukarıdaki örneği, break komutuyla yapmak istersek aşağıdaki gibi yapabiliriz.

flowchart-break-statement.webp

````python
while True:
  cevap = input("3 + 2 = ? \n")
  if cevap == "5":
    break

print("Döngü bitti")
````

## Continue Komutu
Bir döngü içinde bazı koşullar gerçekleştiğinde, döngünün belirli bir dönümünü atlamak için kullanılır.



Continue-statement-python1.jpg

### Örnek:
5 arkadaş, beraber partiye gidecekken Tarık'ın işi çıkıyor ve partiye gelmeyeceğini belirtiyor.
Bu durumu aşağıdaki gibi belirtebiliriz.

````python
arkadaslar = ["Ali", "Ahmet", "Tarık", "Melis", "Bilal"] # 5 tane arkadaş

for isim in arkadaslar:
  if isim == "Tarık":
    continue
  print(isim," partiye geliyor.")
````
