# Koşul ifadeleri ve if yapıları
Bir koşul ifadesi esasında, doğru/yanlış(ingilizce adlarıyla True/False) şeklinde bir bilgidir. Bir koşul oluşturmak için 2 adet değer (ya da değişken) ve bir karşılaştırma operatörüne ihtiyacımız vardır. 
## Karşılaştırma operatörleri
Bir karşılaştırma operatörü solundaki değeri sağındaki ile karşılaştırır.
Python programlama dilinde 6 tana temel  karşılaştırma operatörü bulunur:

```
<   # Küçüktür.
<=  # küçük yada eşittir (büyük değildir)
>   # Büyüktür.
>=  # Büyük yada eşittir (küçük değildir)
==  # Eşittir ( tek '=' işareti bir karşılaştırma operatörü değildir.)
!=  # Eşit değildir.
```
Bu operatörler sayıları ya da stringleri karşılaştırmak için kullanılabilir.Lakin, bir sayı ile bir string'i karşılaştırmak isterseniz hata verecektir.
İki string'i karşılaştırmak istendiğinde bildiğimiz sözlükteki sıralamaya uygun olarak baştaki harfin alfabedeki yerine bakar. (Mesela ('arı' < 'balon' ) ifadesi doğru cevabını döndürür)


````python
('arı' < 'balon')
````

````python
('balon' > 5)
````

# If yapıları
If yapıları, programlamada koşulların sonuçlarına göre bazı kod parçalarının çalışmasını sağlar. 
## If ifadesi
Ingilizcede 'eğer' anlamına gelmektedir. bir koşulun doğru olduğu takdirde çalıştırılacak kodları içerir.
Bir if yapısının 2 bölümü vardır:

*   koşul
*   koşul doğruysa çalışacak olan kod

Bir if ifadesi şu şekilde yazılır:


```
if koşul ifadesi:
    kod satırı1
    kod satırı2
    kod satırı3
```
Görüldüğü gibi; önce if kelimesi kullanılır, ardından koşul belirtilir ve koşulun sonuna : işareti konur, sonrasında da alt satırda bir miktar boşluk bırakılır.(genelde 2 ya da 4 boşluk(tab tuşuna bir vuruş) bırakılır ama önemli olan ka boşluk bırakıldığı değil, sürekli aynı sayıda boşluk bırakılıp bırakılmadığıdır.If ifadesinin kapsamından çıkmak için boşluk bırakmadan if kelimesiyle aynı hizada kod yazabilirsiniz.

````python
if 6 < 5:
    print("4 5'den küçüktür!")
    print("5 4'den büyüktür!")
print("koşul yapısından çıkış")
````

eğer ifade yanlışsa sonra gelen kod parçalarını yok sayacaktır.

````python
if 3 < 2:
  print("3, 2'den küçüktür!")
````

## Else Ifadesi
Bir if ifadesinden sonra kullanılabilen ve if ifadesindeki koşulun yanlış olması durumunda ne yapılacağının belirtildiği ifadedir.

else ifadesinin kullanımı if'e benzer
else kelimesini ve : işaretini koyduktan sonra bir alt satırda boşluk bırakarak çalışacak kodlar yazılır.

````python
if 2 > 3:
  print("2 3'den büyüktür.")
else:
  print("2 3'e eşit ya da 3'ten küçüktür.")
````

## Elif Ifadesi
Else ve if kelimelerinin birleşiminden meydana gelir. Bir üstündeki koşulun yanlış olduğu takdirde başka bir koşul test edilir.

````python
### kullanıcıdan bir sayı isteyen ve bu sayının 5'den büyük olup olmadığını test eden program
### Dikkat : aşağıdaki program, kullanıcı bir sayı girmezse hata verecektir.
### İleriki derslerde programın nasıl hata vermeden çalışacağını öğreneceğiz lakin şuan kullanıcı bir sayı girmek zorunda
sayi = int(input("Bir sayı giriniz : "))
if sayi < 5:
  print("Girdiğiniz sayı 5'den küçük.")
elif sayi == 5:
  print("Girdiğiniz sayı : 5")
else:
  print("Girdiğiniz sayı 5'den büyük.")
````

# İçiçe if ifadeleri
Bir if ifadesinin içinde yeni if ifadeleri de kullanabiliriz. Yeni if ifadeleri yazılırken önceki if ifadesinden dolayı kullanılan boşluklar ekstra olarak kullanılır


```
if koşul 1:
    if koşul 2:
        if koşul 3:
            koşul 3 için kod satırları
            koşul 3 için kod satırları
            koşul 3 için kod satırları
        koşul 2 için kod satırları
    koşul 1 için kod satırları       
```



````python
### 1 ile 8 arasında akılda tutulan rakamı tahmin eden program
cevap1 = input("aklınızdaki sayı 4'den büyük mü?")

if cevap1.lower() == "evet":
  if input("aklınızdaki sayı 6'dan büyük mü?"):
    if input("aklınızdaki sayı 7'den büyük mü?"):
      print("aklınızdaki sayı : 8")
    else:
      print("aklınızdaki sayı : 7")
  elif input("aklınızdaki sayı 5'den büyük mü?").lower() == "evet":
    print("aklınızdaki sayı : 6")
  else:
      print("aklınızdaki sayı : 5")
else:
  if input("aklınızdaki sayı 2'dan büyük mü?").lower() == "evet":
    if input("aklınızdaki sayı 3'den büyük mü?").lower() == "evet":
      print("aklınızdaki sayı : 4")
    else:
      print("aklınızdaki sayı : 3")
  elif input("aklınızdaki sayı 1'den büyük mü?").lower() == "evet":
    print("aklınızdaki sayı : 2")
  else:
      print("aklınızdaki sayı : 1")
````
